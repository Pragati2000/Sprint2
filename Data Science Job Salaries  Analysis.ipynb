# **Data Visualization**

## Visualization No.1 
  **Job Title: The title or position of the job.**

```python
df['Job_Title'].nunique()
```

```python
top_10_job_title = df['Job_Title'].value_counts()[:10]
top_10_job_title
```

```python
plt.figure(figsize = (16, 7))
top_10_job_title.sort_values(ascending = True).plot(kind = 'barh', color = sns.color_palette('tab10'), edgecolor = 'black')
plt.ylabel('Job Titles', fontsize = 16)
plt.xlabel('\nNumber of Occurrences', fontsize = 16)
plt.title('Top 10 Job Titles\n', fontsize = 16)
plt.xticks(rotation = 0, ha = 'center', fontsize = 16)
plt.tight_layout()
plt.show()
```

**Insight**   
In above plot we can see that Data Scientist have maximum number of Occurrences.

## Visualization No.2 
  **Employment Type: The type of employment, such as full-time, part-time, contract.**

```python
df["Employment_Type"].value_counts()
```

```python
plt.figure(figsize = (20, 6))
colors = ['orange', 'green', 'yellow']
ax = df["Employment_Type"].value_counts().plot(kind = 'bar', color = colors, rot = 0)
ax.set_xticklabels(('Full-Time', 'Contract', 'Part-Time'))

for p in ax.patches:
    ax.annotate(int(p.get_height()), (p.get_x() + 0.25, p.get_height() + 1), ha = 'center', va = 'bottom', color = 'black')
    ax.tick_params(axis = 'both', labelsize = 15)
plt.xlabel('Employment Type', weight = "bold", color = "#D71313", fontsize = 14, labelpad = 20)
plt.ylabel('Number of Occurrences', weight = "bold", color = "#D71313", fontsize = 14, labelpad = 20);
```

**Insight**  
  Full-Time: 94 individuals are employed in a standard, full-time capacity, typically working 35-40 hours per week.   
  Contract: 1 individuals are engaged under a contractual agreement, usually for specific projects or defined time periods.   
  Part-Time: 1 individuals work fewer hours than full-time employees, often with flexible schedules and limited benefits.

## Visualization No.4

## Visualization No.5
  **Company Size: The size of the company, often categorized by the number of employees.**

```python
# Convert value counts of "Company Size" column into percentages
company_size_percentages = df["Company_Size"].value_counts(normalize=True) * 100

# Display the result
print(company_size_percentages)
```

```python
plt.figure(figsize = (20, 6))

# Define labels and colors
labels = ['Medium', 'Large', 'Small']  # Match the desired order
colors = ['#FF9999', '#66B2FF', '#99FF99']  # Example color palette

# Create the pie chart
plt.figure(figsize=(5, 5))
plt.pie( company_size_percentages,  labels=labels,  autopct='%1.1f%%', startangle=90, colors=colors, )

# Add a title
plt.title('Company Size Distribution', fontsize=10, weight='bold', color="#D71313")

# Show the pie chart
plt.show()
```

**Insight**  
  Medium: There are 67.7% individuals working in medium-sized companies, which often indicates a substantial workforce and organizational scale.  
  Large: 22.9% individuals are employed by large companies, suggesting significant corporate entities with extensive resources.  
  Small: 9.4% individuals are part of small companies, typically characterized by a more intimate work environment and fewer employees.  

## Visualization No.6
  **Salary Currency: The location of the company, often categorized by the Salary Currency.**

```python
df["Salary_Currency"].value_counts()
```

```python
plt.figure(figsize=(20, 6))
currency_counts = df["Salary_Currency"].value_counts()

top_count = 3
top_currencies = currency_counts.head(top_count)

other_count = currency_counts.shape[0] - top_count
other_sum = currency_counts.iloc[top_count:].sum()
top_currencies['Others'] = other_sum

colors = ["#6528F7", "#F1C93B", "#C8FFE0", "#FF52A2", "#F48484", "#F0DE36"]

explode = [0.1 if i == "Others" else 0 for i in top_currencies.index]

top_currencies.plot(kind = 'pie',fontsize = 12,colors = colors,explode = explode,autopct = '%1.1f%%')
plt.xlabel('Salary Currency', weight="bold", color="#2F0F5D", fontsize=14, labelpad=20)
plt.axis('equal')
plt.legend(labels=top_currencies.index, loc="best")
plt.show()
```

**Insight**  
  In above pie chart we can see that United States Dollar maximum % of Salary Currency.
